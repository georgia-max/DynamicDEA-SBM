CRS  (constant returns-to-scale)

```python
##########################  CRS_dual   ############ 
## CRS primal model Output  oriented
from gurobipy import*


def CRS_output_oriented(X,Y): 
  Output=pd.DataFrame()
  I=X.shape[1]
  O=Y.shape[1]

  for r in range(X.shape[0]):
    model = Model("CRS_model")
    model.setParam('OutputFlag', False)  # Muting the optimize function
    v, u = {}, {}

    for i in range(I):
        v[r, i] = model.addVar(vtype=GRB.CONTINUOUS, name='v_%s%d' % (r, i), lb=0.0001)
    for j in range(O):
        u[r, j] = model.addVar(vtype=GRB.CONTINUOUS, name='u_%s%d' % (r, j), lb=0.0001)

    model.update()

    obj = quicksum(u[r, j] * Y[j][r] for j in range(O))
    model.setObjective(obj, GRB.MAXIMIZE)

    model.addConstr(quicksum(v[r, i] * X[i][r] for i in range(I)) == 1)

    for k in range(X.shape[0]):
        model.addConstr(
            quicksum(u[r, j] * Y[j][k] for j in range(O)) - quicksum(v[r, i] * X[i][k] for i in range(I)) <= 0)

    model.optimize()
    Output.at[r,'Efficiency Score']=model.objVal
  return  Output

```

CRS prim  dual

```python
def CRS_Dual(X,Y): 
  Output=pd.DataFrame()
  I=X.shape[1]
  O=Y.shape[1]
  for r in range(X.shape[0]):
    theta, lamd = {}, {}

    # Initialize LP model
    m = Model("Dual_of_CRS_model")
    m.setParam('OutputFlag', False)  # Muting the optimize function

    # Add decision variables
    for k in range(X.shape[0]):
        lamd[k] = m.addVar(vtype=GRB.CONTINUOUS, name="λ_%s" % k) #k*r times
    theta[r] = m.addVar(vtype=GRB.CONTINUOUS, lb=-1000, name="theta_%s" % r) #r times
    
    m.update()
    # Add objective function
    m.setObjective(theta[r], GRB.MINIMIZE)
    # Add constraints
    for i in range(I):
        m.addConstr(quicksum(lamd[k] * X[i][k] for k in range(X.shape[0])) <= theta[r] * X[i][r])
    for j in range(O):
        m.addConstr(quicksum(lamd[k] * Y[j][k] for k in range(X.shape[0])) >= Y[j][r])

    # Start optimize the formulation
    m.optimize()
    Output.at[r,'Efficiency Score']= m.objVal
    nn=0
    for i in m.getVars():
        if nn <5:  Output.at[r,'Lambdas '+'%s'%nn]= i.x 
        else:      Output.at[r,'Thetha '+'%s'%(nn-5)]= i.x 
        nn=nn+1 

    nn=0
    for c in m.getConstrs():
         Output.at[r,'Slacks '+'%s'%nn]=c.Slack
         nn=nn+1  
  return Output
```

SBM-I-C-dual

```python
#########################  SBM-I-C-dual  ###################################
## SBM- Output-CRS-Primal
# 2 inputs and 2 outputs
## CRS primal model Output  oriented
# 2 inputs and 2 outputs
#data is reference to chapter 2 p.32,
# Objective function: minimize

from gurobipy import*
def CRS_primal_Output_oriented(X,Y):
  Output=pd.DataFrame()
  T={}
  SO={}
  SI={}
  m=X.shape[1] # numbers of Input
  s=Y.shape[1]
  for o in range(X.shape[0]):

    lamd = {}
    spos ={}
    t={}
    # Initialize LP model
    model = Model("Primal_SBM_output")
    model.setParam('OutputFlag', False)  # Muting the optimize function

    # Add decision variables
    for j in range(X.shape[0]):
        lamd[j] = model.addVar(vtype=GRB.CONTINUOUS, ub=2, name="λ%0.3s:" % j)

    # output variable S_positive
    for r in range(s):
        spos[r] = model.addVar(vtype=GRB.CONTINUOUS, name ="SO_%s:" % r)

    model.update()

    # Add objective function
    model.setObjective(1+ (1 / s) * (quicksum(spos[r] / Y[r][o] for r in range(s))), GRB.MAXIMIZE)

     #input contraint
    for i in range(m):
        model.addConstr(quicksum(lamd[j] * X[i][j] for j in range(X.shape[0])) <= X[i][o])

     #output constraint
    for r in range(s):
        model.addConstr(quicksum(lamd[j] * Y[r][j] for j in range(X.shape[0])) - spos[r] == Y[r][o])

    # Start optimize the formulation
    model.update()
    model.optimize()
    Output.at[o,'Efficiency Score']=1/model.objVal
    for j in range(X.shape[0]):
        Output.at[o,'Lambdas %s'%j]= lamd[j].x
    for r in range(s):
        Output.at[o,'SO %s'%r]= spos[r].x    

  return Output
```

SBM-Non


```python
############################   SBM-Non #########################
## SBM- Non-CRS-Primal
# 2 inputs and 2 outputs

#data is reference to tone(2001)
## georgia Liu Mar22

from gurobipy import*
def CRS_non_primal_Output_oriented(X,Y):
  Output=pd.DataFrame()
  T={}
  SO={}
  SI={}
  m=X.shape[1] # numbers of Input
  s=Y.shape[1]
  for o in range(X.shape[0]):

    lamd = {}
    spos, sneg = {},{}
    t={}
    # Initialize LP model
    model = Model("Primal_SBM_output")
    model.setParam('OutputFlag', False)  # Muting the optimize function

    # Add decision variables
    for j in range(X.shape[0]):
        lamd[j] = model.addVar(vtype=GRB.CONTINUOUS, ub=2, name="λ%0.3s:" % j)

    t = model.addVar(vtype=GRB.CONTINUOUS, name="t:")

    # output variable S_positive
    for r in range(s):
        spos[r] = model.addVar(vtype=GRB.CONTINUOUS, name ="SO_%s:" % r)

    #input variable S_negative
    for i in range(m):
        sneg[i] = model.addVar(vtype=GRB.CONTINUOUS, name="SI_%s:" % i)
    
    model.update()

    # Add objective function
    model.setObjective(t - (1 / m) * (quicksum(sneg[i] / X[i][o] for i in range(m))), GRB.MINIMIZE)

    #Add constraints
    model.addConstr(t + (1 / s) * (quicksum(spos[r] / Y[r][o] for r in range(s))) == 1)


     #input contraint
    for i in range(m):
        model.addConstr(quicksum(lamd[j] * X[i][j] for j in range(X.shape[0])) + sneg[i] == t * X[i][o])

     #output constraint
    for r in range(s):
        model.addConstr(quicksum(lamd[j] * Y[r][j] for j in range(X.shape[0])) - spos[r] == t*Y[r][o])

    # Start optimize the formulation
    model.update()
    model.optimize()
    Output.at[o,'Efficiency Score']=model.objVal
    Output.at[o,'T value']=t.x

    for j in range(X.shape[0]):
        Output.at[o,'Lambdas %s'%j]= lamd[j].x / t.x
    for r in range(s):
        Output.at[o,'SO %s'%r]= spos[r].x  / t.x 
    for i in range(m):
        Output.at[o,'SI %s'%i]= sneg[i].x  / t.x  

  return Output
```

VRS_Model: (Variable returns to scale):


https://www.gurobi.com/documentation/9.1/refman/java_model_addvar.html

```python
def VRS_model(X,Y): 
  E=[] 
  UU=[]  
  I=X.shape[1]
  O=Y.shape[1]
  for r in range(X.shape[0]):            
    m=Model("VRS_model")
    m.setParam('OutputFlag', False)  # Muting the optimize function
    v,u,u0={},{},{}
    for i in range(I):
        v[r,i]=m.addVar(vtype=GRB.CONTINUOUS,name="v_%d%d"%(r,i))
    for j in range(O):
        u[r,j]=m.addVar(vtype=GRB.CONTINUOUS,name="u_%d%d"%(r,j))
    u0[r]=m.addVar(lb=-1000,vtype=GRB.CONTINUOUS,name="u_0%d"%r)
    m.update()
    m.setObjective(quicksum(u[r,j]*Y[j][r] for j in range(O))-u0[r],GRB.MAXIMIZE)
    m.addConstr(quicksum(v[r,i]*X[i][r] for i in range(I))==1)
    for k in range(X.shape[0]):
        m.addConstr(quicksum(u[r,j]*Y[j][k] for j in range(O)) -quicksum(v[r,i]*X[i][k] for i in range(I))-u0[r] <=0)
    m.optimize()
    E.append(m.objVal)  
    UU.append(u0[r].X)

  Output_VRC=pd.DataFrame()
  Output_VRC['Efficiency Score']=E
  Output_VRC['U(r) value']=UU
  return Output_VRC
```

```python
X=pd.DataFrame()
X[0]=[11,7,11,14,14]
X[1]=[14,7,14,14,15]
Y=pd.DataFrame()
Y[0]=[2,1,1,2,3]
Y[1]=[2,1,1,3,2]
Y[2]=[1,1,2,1,3]
VRS_model(X,Y)
```

```python
#######################  NSBM-input  ###########################
"""Tone, Tsutsui (2009) Network DEA paper"""
from gurobipy import*

def Network_SBM(N_X,N_Y,Link,Weight_node,VRS,Link_case,orientation):  #0)return_to_scale:'yes' or 'no'  1) Link_case: 'fixed' or 'free'    2) orientation: 'input' or 'output' 'non'
  Output=pd.DataFrame()
  E_div1,E_div2,E_div3,E_total ={},{},{},{}
  SO,SI={},{}

  for o in range(N_X[0].shape[0]):
    N=len(N_X) 

    "This data is from Tone, Tsutsui (2009) Network DEA paper"
    lamd ={};lamd_X={};lamd_Y={};lamd_link={}
    model = Model("NSBSM-Nonoriented ")
    model.setParam('OutputFlag', False)  # Muting the optimize function
    #Add decision variables
    for k in range(N):
        for j in range(N_X[0].shape[0]):
            lamd[k,j] = model.addVar(vtype=GRB.CONTINUOUS, name="λ_%s_%s:" % (k,j))
  


    # output
    for k in range(N):
      if type(N_Y[k]) == pd.DataFrame: 
            for r in range(N_Y[k].shape[1]):
                SO[k,r] = model.addVar(vtype=GRB.CONTINUOUS, name="SO%s_%s:" % (k,r))
  
    #input
    for k in range(N):
      if type(N_X[k]) == pd.DataFrame: 
            for i in range(N_X[k].shape[1]):
                SI[k,i] = model.addVar(vtype=GRB.CONTINUOUS, name="SI%s_%s:" % (k,i))


    if orientation=='non':
      #T={} 
      #t = model.addVar(vtype=GRB.CONTINUOUS, name="t:")
      t=1 
    else: t=1

    model.update()
  

    objective_XX=0
    objective_YY=0
    
    if (orientation=='input'):
      for r in range(N):
        if type(N_X[r]) == pd.DataFrame: 
                 objective_XX=(Weight_node[r]* (t - (1/N_X[r].shape[1])*   quicksum(   (SI[r,p]/N_X[r][p][o])  for p in range(N_X[r].shape[1]) )     )  )+objective_XX
      model.setObjective(    objective_XX      , GRB.MINIMIZE)  
    elif  (orientation=='ouput'): 
      for r in range(N):
        if  type(N_Y[r]) == pd.DataFrame:
                 objective_YY=(Weight_node[r]* (1 + (1/N_Y[r].shape[1])*   quicksum(   (SO[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )     )  )+objective_YY   
      model.setObjective(    objective_YY      , GRB.MAXIMIZE)
    elif (orientation=='non'):
      for r in range(N):
        if type(N_X[r]) == pd.DataFrame: 
                 objective_XX=(Weight_node[r]* (t - (1/N_X[r].shape[1])*   quicksum(   (SI[r,p]/N_X[r][p][o])  for p in range(N_X[r].shape[1]) )     )  )+objective_XX
        if  type(N_Y[r]) == pd.DataFrame:
                 objective_YY=(Weight_node[r]* (1 + (1/N_Y[r].shape[1])*   quicksum(   (SO[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )     )  )+objective_YY    
      model.setObjective(    objective_XX-objective_YY      , GRB.MINIMIZE)
    

    #if orientation=='non':
    #     XX=0
    #     total_Weight_node=0
    #     for r in range(N):
    #       if type(N_Y[r]) == pd.DataFrame: 
    #             XX=(Weight_node[r]* (1 + (1/N_Y[r].shape[1])*   quicksum(   (SO[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )     )  )+XX 
    #             total_Weight_node=Weight_node[r]+total_Weight_node
    #     model.addConstr(XX ==total_Weight_node)


    #model.setObjective(    quicksum( (Weight_node[r]* (t - (1/N_X[r].shape[1])*   quicksum(   (SI[r,p]/N_X[r][p][o])  for p in range(N_X[r].shape[1]) )     )  )    for r in range(N) )
    #                   , GRB.MINIMIZE)

    for r in range(N):  
      if type(N_X[r]) == pd.DataFrame: 
        for p in range(N_X[r].shape[1]): 
           model.addConstr(quicksum(lamd[r,j] * N_X[r][p][j] for j in range(N_X[0].shape[0])) + SI[r,p] ==  t*N_X[r][p][o])
      if type(N_Y[r]) == pd.DataFrame:
        for p in range(N_Y[r].shape[1]):    
           model.addConstr(quicksum(lamd[r,j] * N_Y[r][p][j] for j in range(N_X[0].shape[0])) - SO[r,p] ==  t*N_Y[r][p][o])
    




    if Link_case=='free':
      for ii in range(len(Link)):
        for jj in range(len(Link)):
          if type(Link[ii][jj]) == pd.DataFrame : 
            for p in range(Link[ii][jj].shape[1]): 
               model.addConstr(quicksum(Link[ii][jj][p][j] *lamd[ii,j] for j in range(N_X[0].shape[0])) == quicksum(Link[ii][jj][p][j] *lamd[jj,j] for j in range(N_X[0].shape[0])))


    if Link_case=='fixed':
      for ii in range(len(Link)):
        for jj in range(len(Link)):
          if type(Link[ii][jj]) == pd.DataFrame : 
            for p in range(Link[ii][jj].shape[1]): 
               model.addConstr(Link[ii][jj][p][o] == quicksum(Link[ii][jj][p][j] * lamd[ii,j] for j in range(N_X[0].shape[0])))
               model.addConstr(Link[ii][jj][p][o] == quicksum(Link[ii][jj][p][j] * lamd[jj,j] for j in range(N_X[0].shape[0])))


    if VRS=='yes':
      for r in range(N):  
            model.addConstr(quicksum(lamd[r,j] for j in range(N_X[0].shape[0])) == 1)


    model.update()
    model.optimize()
    


    #for k in range(N):
    #    for j in range(X.shape[0]):
    #       Output.at[o,'Lambdas'+'node%s'%k+',%s'%j]= lamd[k,j].x     
    #
    SO_sol= model.getAttr('x',SO )
    # print("SO_sol",SO_sol)
    SI_sol = model.getAttr('x',SI)
    # print("SI_sol",SI_sol)

    if (orientation=='input'): Output.at[o,'Efficiency Score']=model.objVal
    elif (orientation=='output'): 
       YY=0; n=0
       for r in range(N):
           if  type(N_Y[r]) == pd.DataFrame:
             YY=(1/ ( (1 + (1/N_Y[r].shape[1])*   sum(   (SO_sol[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )   )  )  )+YY ;n=n+1
       Output.at[o,'Efficiency Score']=YY/n     
   
    elif (orientation=='non'): 
      A=0;B=0;C=0 
      for r in range(N):
           if  type(N_X[r]) == pd.DataFrame: A=(1 - (1/N_X[r].shape[1])*   sum(   (SI_sol[r,p]/N_X[r][p][o])  for p in range(N_X[r].shape[1]) )     )     
           else: A=1
           if  type(N_Y[r]) == pd.DataFrame: B=(1/ ( (1 + (1/N_Y[r].shape[1])*   sum(   (SO_sol[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )   )  )  )
           else: B=1            
           C=(A*B)+C     
      Output.at[o,'Efficiency Score']=C/N


      #objective_XX=0
      #objective_YY=0
      #for r in range(N):
      #  if type(N_X[r]) == pd.DataFrame: 
      #           objective_XX=(Weight_node[r]* (t - (1/N_X[r].shape[1])*   sum(   (SI_sol[r,p]/N_X[r][p][o])  for p in range(N_X[r].shape[1]) )     )  )+objective_XX
      #  if  type(N_Y[r]) == pd.DataFrame:
      #           objective_YY=(Weight_node[r]* (1 + (1/N_Y[r].shape[1])*   sum(   (SO_sol[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )     )  )+objective_YY  
      #Output.at[o,'Efficiency Score']=objective_XX/objective_YY 

    #if orientation=='non': Output.at[o,'T value']=t.x

    for r in range(N): 
      if orientation=='input':
        if  type(N_X[r]) == pd.DataFrame:
          Output.at[o,'Effinecy Node'+str(r+1) ]="%0.3f" %(1 - (1/N_X[r].shape[1])*   sum(   (SI_sol[r,p]/N_X[r][p][o])  for p in range(N_X[r].shape[1]) )     )     
      if orientation=='output':
        if  type(N_Y[r]) == pd.DataFrame:
          Output.at[o,'Effinecy Node'+str(r+1) ]="%0.3f" % (1/ ( (1 + (1/N_Y[r].shape[1])*   sum(   (SO_sol[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )   )  )  )
          #Output.at[o,'Effinecy Node'+str(r+1) ]=1/int(Output['Effinecy Node'+str(r+1) ][o])
      elif orientation=='non':
           if  type(N_X[r]) == pd.DataFrame: A=(1 - (1/N_X[r].shape[1])*   sum(   (SI_sol[r,p]/N_X[r][p][o])  for p in range(N_X[r].shape[1]) )     )     
           else: A=1
           if  type(N_Y[r]) == pd.DataFrame: B=(1/ ( (1 + (1/N_Y[r].shape[1])*   sum(   (SO_sol[r,p]/N_Y[r][p][o])  for p in range(N_Y[r].shape[1]) )   )  )  )
           else: B=1            
           Output.at[o,'Effinecy Node'+str(r+1) ]="%0.3f" % (A*B)     

 
  


    ################################  Slack   #########################    
    for r in range(N):  
      if type(N_X[r]) == pd.DataFrame:
        if (orientation=='input') or  (orientation=='output'):
           for i in range(N_X[r].shape[1]): Output.at[o,'SI'+str(r+1)+'%s'%i]=SI[r, i].x 
        elif orientation=='non': 
           for i in range(N_X[r].shape[1]): Output.at[o,'SI'+str(r+1)+'%s'%i]=SI[r, i].x / t
    for r in range(N): 
      if type(N_Y[r]) == pd.DataFrame:
        if (orientation=='input') or (orientation=='output'):
          for i in range(N_Y[r].shape[1]):Output.at[o,'SO'+str(r+1)+'%s'%i]=SO[r, i].x  
        elif orientation=='non': 
          for i in range(N_Y[r].shape[1]):Output.at[o,'SO'+str(r+1)+'%s'%i]=SO[r, i].x/ t   

    ################################  Target   ######################### 
    for r in range(N):  
      if type(N_X[r]) == pd.DataFrame:
        if (orientation=='input') or  (orientation=='output'):
           for i in range(N_X[r].shape[1]): Output.at[o,'target_X'+str(r+1)+'%s'%i]=N_X[r][i][o]-SI[r, i].x 
        elif orientation=='non': 
           for i in range(N_X[r].shape[1]): Output.at[o,'target_X'+str(r+1)+'%s'%i]=N_X[r][i][o]-SI[r, i].x  / t 

    for r in range(N): 
      if type(N_Y[r]) == pd.DataFrame:
        if (orientation=='input') or (orientation=='output'):
          for i in range(N_Y[r].shape[1]): Output.at[o,'target_Y'+str(r+1)+'%s'%i]=N_Y[r][i][o]+SO[r, i].x   
        elif orientation=='non': 
          for i in range(N_Y[r].shape[1]): Output.at[o,'target_Y'+str(r+1)+'%s'%i]=N_Y[r][i][o]+SO[r, i].x / t 

  return Output


```

```python
def DEA_outputs(N_X,N_Y,Link,Weight_node,VRS,Link_case,orientation):

  bold_start = '\033[1m'
  bold_end   = '\033[0m'
  if len(N_X) > 1: 
    print(bold_start,'Results of network DEA model:')
    display(Network_SBM(N_X,N_Y,Link,Weight_node,VRS,Link_case,orientation))
    ######################   Inputs from network DEA #######################
    AA=pd.concat(N_X,axis=1)
    AA.set_axis(range(len(N_X)), axis=1, inplace=True)
    ######################   Outputs from network DEA #######################
    BB=pd.DataFrame()
    X=[BB]
    for i in range(len(N_Y)):
      if type(N_Y[i]) == pd.DataFrame:
         X.append(N_Y[i])
    BB=pd.concat(X,axis=1)   
    BB.set_axis(range(BB.shape[1]), axis=1, inplace=True)
    #########################################################################
    ##############################################################
    X,Y=AA,BB
  else: 
    X,Y=N_X[0],N_Y[0]
  
  print(bold_start,'CRS  (constant returns-to-scale):')
  display(CRS_output_oriented(X,Y))
  print(bold_start,'CRS prim  dual:')
  display(CRS_Dual(X,Y))
  print(bold_start,'SBM-I-C-dual:')
  display(CRS_primal_Output_oriented(X,Y))
  print(bold_start,'SBM-Non:')
  display(CRS_non_primal_Output_oriented(X,Y))
  print(bold_start,'VRS_Model: (Variable returns to scale):')
  display(VRS_model(X,Y))
```
